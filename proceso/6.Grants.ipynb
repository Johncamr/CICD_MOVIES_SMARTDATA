## Asignacion de Permisos por Grupo de acuerdo a los roles seleccionados

```python
%sql
GRANT USE CATALOG ON CATALOG catalog_movie_project TO `DEs`;
GRANT USE CATALOG ON CATALOG catalog_movie_project TO `data_engineers`;
GRANT USE CATALOG ON CATALOG catalog_movie_project TO `end_users`;
```

```python
%sql
GRANT USE SCHEMA ON SCHEMA  catalog_movie_project.bronze TO `data_engineers`;
GRANT SELECT  ON SCHEMA catalog_movie_project.bronze TO `data_engineers`;
GRANT ALL PRIVILEGES  ON SCHEMA catalog_movie_project.bronze TO `data_engineers`;

GRANT USE SCHEMA ON SCHEMA  catalog_movie_project.silver TO `data_engineers`;
GRANT SELECT  ON SCHEMA catalog_movie_project.silver TO `data_engineers`;
GRANT ALL PRIVILEGES  ON SCHEMA catalog_movie_project.silver TO `data_engineers`;

GRANT USE SCHEMA ON SCHEMA  catalog_movie_project.golden TO `data_engineers`;
GRANT SELECT  ON SCHEMA catalog_movie_project.golden TO `data_engineers`;
GRANT ALL PRIVILEGES  ON SCHEMA catalog_movie_project.golden TO `data_engineers`;
```

```python
%sql
GRANT USE SCHEMA ON SCHEMA  catalog_movie_project.bronze TO `DEs`;
GRANT SELECT  ON SCHEMA catalog_movie_project.bronze TO `DEs`;
GRANT ALL PRIVILEGES  ON SCHEMA catalog_movie_project.bronze TO `DEs`;

GRANT USE SCHEMA ON SCHEMA  catalog_movie_project.silver TO `DEs`;
GRANT SELECT  ON SCHEMA catalog_movie_project.silver TO `DEs`;
GRANT MODIFY  ON SCHEMA catalog_movie_project.silver TO `DEs`;

GRANT USE SCHEMA ON SCHEMA  catalog_movie_project.golden TO `DEs`;
GRANT SELECT  ON SCHEMA catalog_movie_project.golden TO `DEs`;
```
